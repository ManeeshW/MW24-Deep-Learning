```python
# Import PyTorch and torchviz
import torch
import torch.nn as nn
import torchviz

# Define the network class
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        # Define the layers
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        # Define the forward pass
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        return x

# Create an instance of the network
net = SimpleNN(3, 5, 2)

# Create a dummy input
x = torch.randn(1, 3)

# Compute the output and the graph
y = net(x)
graph = torchviz.make_dot(y, params=dict(net.named_parameters()))

# Print the output and the graph
print(y)
graph
```
